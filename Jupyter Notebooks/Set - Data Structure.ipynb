<h3>Declaration of Set</h3>

```python
# Empty set declaration
# myset = {} can not be given in order to create set
# myset = {} - will be considered as dictionary

myEmptySet = set()

# Set intialization
mySet = {"apple", "banana", "cherry", "apple"}
print(mySet)
```

<h3>To initialize a set</h3>

```python
# Set initialization
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = {True, False, False}
set4 = {"abc", 34, True, 40, "male"}

print(set1)
print(set2)
print(set3)
print(set4)
```

<h3>To find Length of the set</h3>

```python
len(mySet)
```

```python
# Use set() constructor to create a set
# Double brackets are required
set5 = set(("apple", "banana", "cherry"))
set5
```

<h3>To access elements in set</h3>

```python
# You cannot access items in a set by referring to an index or a key.
# Access element with for loop
mySet = {"apple", "banana", "cherry","orange"}

for x in mySet:
    print(x)
```

```python
# To check whether the Set contains the element or not?
print("banana" in mySet)
```

<h3>To add items to a Set</h3>

```python
# To add item to a set, we can use add method
mySet2 = {"a","b","c"}
mySet2.add("d")
print(mySet2)
```

<h3>To add set to a set</h3>

```python
# To add a set to a set, we can use update method
set1 = {1,2,3,4,5}
set2 = {"a","b","c"}

set1.update(set2)
print(set1)
```

```python
# update method can accecpt any type of iterable - list, tuple, 
# dictionary, set.
set3 = {1,2,3,4,5}
set4 = ["a","b","c"]

set3.update(set4)
print(set3)
```

<h3>To delete/remove a item from set</h3>

```python
# There are two method avaiable : remove() and discard()
# remove method - will raise an error when there is no specified 
# item available in the set.
set1 = {1,2,3,4,5}
set1.remove(3)  # value 3 removed
#set1.remove(3)  # now what will happen?
set1
```

```python
# discard method - will not raise an error when there is no specified 
# item available in the set.
set1 = {1,2,3,4,5}
set1.discard(3)
set1.discard(3)
set1
```

```python
# To remove the last element from the set
# we can use pop() method
set1 = {"a","b","c","d","apple","banana","cherry"}
x = set1.pop()
print(x)
print(set1)
```

```python
# To remove entries from set
set2 = {1,2,3,4,5}
set2.clear()
set2
```

```python
# To delete an entier set
# we can use del keyword
set3 = {1,2,3}
del set3
print(set3)
```

<h3>To join or combine sets</h3>

```python
# There are various ways to combine two or more sets
# union method - returns a new set with all items from both sets
# update method - we seen earlier (To add a set or any iterable types)
# both method will exclude the duplicate item.

set1 = {1,2,3,4,5}
set2 = {"a","b","c"}

set3 = set1.union(set2)
print(set3)
```

<h3>To keep only the items present in both sets.</h3>

```python
# intersection_update function
# Keep the items that exist in both set1, and set2
set1 = {1,2,3,4,5}
set2 = {2,4,5,6,7}

set1.intersection_update(set2)
print(set1)
```

```python
# Intersection function
# Return a set of items that exist in both set1, and set2
set1 = {1,2,3,4,5}
set2 = {2,4,5,6,7}

set3 = set1.intersection(set2)
print(set3)
```

<h3>To keep only the items present in both sets without duplicates</h3>

```python
# Keep the items that are not present in both sets
# symmetric_difference_update() - keep only the elements that are NOT 
# present in both sets.

set1 = {1,2,3,4,5}
set2 = {2,4,5,6,7}

set1.symmetric_difference_update(set2)
print(set1)
```

```python
# symmetric_difference() - Return a set of items that does not exist in 
# both set1, and set2.

set1 = {1,2,3,4,5}
set2 = {2,4,5,6,7}

set3 = set1.symmetric_difference(set2)
print(set3)
```

<h3>Set methods</h3>

```python
# isdisjoint() - Return True if no items in set1 is present in set2

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "facebook"}
x = set1.isdisjoint(set2)
print(x)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
x = set1.isdisjoint(set2)
print(x)
```

```python
# issubset() - Returns whether another set contains this set or not

a = {"a", "b", "c"}
b = {"f", "e", "d", "c", "b", "a"}
c = a.issubset(b)
print(c)
```

```python
a = {"a", "b", "c"}
b = {"f", "e", "d", "c", "b"}
c = a.issubset(b)
print(c)
```

```python
# issuperset() - Return True if all items set2 are present in set1

set1 = {"f", "e", "d", "c", "b", "a"}
set2 = {"a", "b", "c"}
x = set1.issuperset(set2)
print(x)
```

```python
x = {"f", "e", "d", "c", "b"}
y = {"a", "b", "c"}
z = x.issuperset(y)
print(z)
```

```python
# difference() - Return a set that contains the items that only exist in set1, 
# and not in set2.

set1= {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
x = set1.difference(set2)
print(x)
print(set1)
```

```python
# difference_update() - Remove the items that exist in both sets

x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.difference_update(y)
print(x)
```
