<center><h2>String Manipulations</h2></center>

```python
# Captialize function

myString = "hello world"
print(myString.capitalize())
```

```python
# casefold function

myString = "Hello World"
print(myString.casefold())
```

```python
# Center function

myString = "hello"
print(myString.center(20))
```

```python
# lJust function

myString = "hello world"
print(myString.ljust(20))
```

```python
# rjust function

myString = "hello world"
print(myString.rjust(20))
```

```python
# split function

myString = "Hello, how are you?"
print(myString.split(' '))
```

```python
# rsplit function

myString = "Hello, how are you?"
print(myString.rsplit(' '))
```

```python
# upper function

myString = "hello"
print(myString.upper())
```

```python
# lower function

myString = "HELLO"
print(myString.lower())
```

```python
# title function

myString = "hello, how are you?"
print(myString.title())
```

```python
# expandtabs function

myString = "Hello,\thow\tare\tyou?"
print(myString.expandtabs(3))
print(myString.expandtabs(10))
```

```python
# count function

myString = "racecar"
print(myString.count('a'))
```

```python
# strip function

myString = "   hello    "
print(myString.strip())
```

```python
# rstrip function

myString ="hello       "
print(myString.rstrip())
```

```python
# lstrip function

myString ="       hello"
print(myString.lstrip())
```

```python
# encode function

txt = "My name is Ståle"
print(txt.encode())
```

```python
# find function

myString = "Hello, how are you?"
print(myString.find('?'))
```

```python
# index function

myString = "Hello, how are you?"
print(myString.find('?'))
```
